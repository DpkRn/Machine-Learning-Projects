```python
#importing all library
!pip install textdistance
import numpy as np
import pandas as pd
import textdistance
import re
from collections import Counter
```

```python
#loading datasets
words=[]
with  open('autocorrect book.txt','r',encoding='utf-8') as f:
  data=f.read()
  data=data.lower()
```

```python
print(data[0:100])
```

```python
#tokenizing of word
word=re.findall('\w+',data)
print(word[0:10])
words+=word
print(words[0:10])
print(len(words))
```

```python
#dectionary with word and frequency
word_freq_dict=Counter(words)
```

```python
word_freq_dict.most_common(10)
```

```python
V=list(set(words))
print(len(V))
Counter(V)
```

```python
#finding probablity of word appearing
probs={}
for k in word_freq_dict.keys():
  probs[k]=word_freq_dict[k]/len(words)
probs
```

```python
#finding similarties of word in vocab
def autocorrect(word):
  word.lower()
  if word in probs:
    print('word is already there',word)
  else:
    similarities=[1-(textdistance.jaccard).distance(w,word) for w in probs.keys()]
    df=pd.DataFrame({'word':list(probs.keys()),'probs':list(probs.values())})
    df['similarities']=similarities
    df=df.sort_values(['similarities','probs'],ascending=False)
    print(df.head(10))
```
