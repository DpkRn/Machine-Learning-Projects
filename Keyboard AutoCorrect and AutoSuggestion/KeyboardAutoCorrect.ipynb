```python
#importing required library
import numpy as np
import pandas as pd
import string
import re
from collections import Counter
```

```python
#importing datasets
words=[]
with open('autocorrect book.txt','r',encoding='utf-8') as file:
  data=file.read()
  data=data.lower()
  words=re.findall('\w+',data)
```

```python
print(words[0:10])
print(len(words))
```

```python
#counting each word frequecy so that i can get probablity of appearing
words_count=Counter(words)
vocab=set(words)
print(words_count)
```

```python
#making dict with word probablity
words_probs={}
words_probs={ word:words_count[word]/len(words) for word in words_count.keys()}
words_probs
```

#making possible combination of word so that we take intersection of vocab word and possible combination

```python

def split(word):
  return [(word[:i],word[i:]) for i in range(len(word)+1)]
print(split("hello"))
```

```python
def delete(word):
  return [left+right[1:] for left,right in split(word) if right]
print(delete('hello'))
```

```python
def swap(word):
  return [l+r[1]+r[0]+r[2:] for l,r in split(word) if len(r)>1]
print(swap("hello"))
```

```python
def replace(word):
  letters=string.ascii_lowercase
  return [l+c+r[1:] for l,r in split(word) if r for c in letters]
print(replace("hello"))
```

```python
def insert(word):
  letters=string.ascii_lowercase
  return [l+c+r[1:] for l,r in split(word) for c in letters]
print(insert("hello"))
```

```python
def level_one_edit(word):
  return set((delete(word)+swap(word)+replace(word)+insert(word)))
print(level_one_edit('load'))
```

```python
def level_two_edit(word):
  return [e2 for e1 in level_one_edit(word) for e2 in level_one_edit(e1)]
print(level_two_edit("love"))
```

```python
#auto correct search bar
def auto_correct_spelling(word,vocab,word_probs):
  if word in vocab:
    return word
  else:
    suggetions= level_one_edit(word) or level_two_edit(word) or [word]
    best_suggestions=[x for x in suggetions if x in vocab]

    return [(w,words_probs[w]) for w in best_suggestions]
```

```python
search_text="sentenec"
guess=auto_correct_spelling(search_text,vocab,words_probs)
print(guess)
```
